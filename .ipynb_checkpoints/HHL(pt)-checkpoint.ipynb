## Preparação de Estado

O primeiro passo para inicialização de qualquer algoritmo
quântico é a importação das bibliotecas necessárias para a
construção e execução do circuito. Essas operações podem
ser visualizadas no boxe abaixo.

```python
from qiskit import QuantumRegister,QuantumCircuit, ClassicalRegister,transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_bloch_multivector, plot_distribution
from qiskit.circuit.library import UnitaryGate
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import expm
```

Em seguida, é necessário a declaração dos registradores a serem utilizados no circuito de acordo com o panorama geral apresentado na seção anterior. Para os $q-c\ket{0}^{\otimes n}$, utilizamos $n = 2$ qubits, enquanto que para o $q-b\ket{0}_b$ e $q-a\ket{0}$ utilizamos $1$ qubit para cada, resultando em $4$ qubits no total. Como as medidas são realizadas no $q-a\ket{0}$ e $q-b\ket{0}_b$, utilizamos $n_c=2$ bits clássicos para a medida.

```python
nc = 2
clock = 2
ancilla = 1
qr_clock1 = QuantumRegister(1, 'clock_1')
qr_clock2 = QuantumRegister(1, 'clock_2')
qr_b = QuantumRegister(1, 'b')
qr_ancilla = QuantumRegister(ancilla,'ancilla')
cl = ClassicalRegister(nc, 'cl')
qc = QuantumCircuit(qr_ancilla, qr_clock1, qr_clock2, qr_b, cl)
```

Feita a inicialização correta do algoritmo, a preparação do estado é realizada aplicando um operador quântico que leve $\ket{0}_b = (1 \ \ \ 0)^T$ para o estado $\ket{1}_b = (0 \ \ \ 1)^T$. Isso pode ser feito simplesmente aplicando a porta lógica $X$

\begin{equation}
    X \ket{0}_b = \ket{1}_b = \ket{b}
\end{equation}

Teremos então:

\begin{equation}
    \ket{\varphi_1} = \ket{1} \ket{00} \ket{0}_{a}
\end{equation}

No algoritmo, essa aplicação é feita especificando a porta e os qubits desejados, conforme boxe abaixo

```python
qc.x(qr_b)
qc.barrier()
```

Cabe destacar que, ao final de cada rotina, aplicamos uma
barreira no circuito, visando torná-lo mais organizado.

## Estimativa de Fase Quântica

Neste ponto, a rotina QPE é aplicada no circuito quântico, iniciando com uma sobreposição uniformemente distribuída nos clock qubits utilizando a porta Hadamard:

```python
qc.h(qr_clock1)
qc.h(qr_clock2)
qc.barrier()
```

Antes de aplicarmos a unitária controlada faremos uma mudança de base para os autovetores base de $A$. Contanto que $\ket{1}=\frac{1}{\sqrt{2}}(-\ket{u_1}+\ket{u_2})$, temos que $b_1=-\frac{1}{\sqrt{2}}$ e $b_2=\frac{1}{\sqrt{2}}$. Logo:

\begin{equation}
    \begin{split}
        \ket{\varphi_2} &= \ket{1} \frac{1}{2}(\ket{00} + \ket{01} + \ket{10} + \ket{11})\ket{0}_{a} \\
        \ket{\varphi_2} &= \frac{1}{\sqrt{2}}(-\ket{u_1}\ket{00}-\ket{u_1}\ket{01}-\ket{u_1}\ket{10}-\ket{u_1}\ket{11} \\
        &+\ket{u_2}\ket{00}+\ket{u_2}\ket{01}+\ket{u_2}\ket{10}+\ket{u_2}\ket{11})
    \end{split}
\end{equation}

O próximo passo é a aplicação do operador $e^{iAt}$ considerando

\begin{equation} \label{eq:seq_U}
    U^l \ket{b} = UUU \dots U \ket{b}= \left( e^{2 \pi i \theta}\right)^l \ket{b} = e^{2 \pi i \theta \cdot l} \ket{b}
\end{equation}

em que cada clock qubit controla uma parte da evolução temporal. Para tanto, utilizamos portas lógicas controladas, de modo que as rotações sejam escolhidas a partir da decomposição espectral da matriz $A$, cuja matriz de transformação de base é escrita como

\begin{equation}
    V = \begin{pmatrix}
\frac{1}{\sqrt{2}} &  \frac{1}{\sqrt{2}} \\
-\frac{1}{\sqrt{2}} &  \frac{1}{\sqrt{2}}\\
\end{pmatrix}
\end{equation}

Nesse contexto, o operador $U$ deve ter sua representação diagonal defina como

\begin{equation} \label{eq:U_d}
    U_{D} = \begin{pmatrix}
    e^{i\lambda_1t} & 0 \\
    0 &  e^{i\lambda_2t}\\
\end{pmatrix}
\end{equation}

Note que a condição do operador $\tilde \lambda_j = \tilde N\lambda_jt/2\pi$ ser inteiro é atendida para determinados valores de $t$. Escolhendo $t = \pi/2$, teremos $\{\tilde \lambda_1 = \lambda_1 =  1, \tilde \lambda_2 = \lambda_2 =2 \}$, sabendo que a representacão diagonal de $U$ é dada por

\begin{equation}
    U_{D} = \begin{pmatrix}
    e^{i\lambda_1t} & 0 \\
    0 &  e^{i\lambda_2t}\\
\end{pmatrix}
\end{equation}

será então reescrita como

\begin{equation}
    U_{d} = \begin{pmatrix}
    i & 0 \\
    0 &  -1\\
    \end{pmatrix}
\end{equation}

Finalmente, podemos escrever $U$ na base de $A$ a partir da matriz de transformação $U = VU_dV^{-1}$, resultando em

\begin{equation} \label{eq:1}
    U = \frac{1}{2}\begin{pmatrix}
    -1 + i & 1+i \\
    1+i &  -1+i\\
    \end{pmatrix}
\end{equation}

O processo para conseguirmos operadores unitários de ordem maior é assim reduzido a $U^l = VU_d^lV^{-1}$. Para $n=2$ c-qubits, $l=2$ é suficiente para fornecer uma precisão aceitável para o algoritmo, de modo que $U^2$ é então escrita como

\begin{equation}
    U^2 = \begin{pmatrix}
    0 & -1 \\
    -1 &  0\\
    \end{pmatrix}
\end{equation}


Na simulação, a aplicação desse operador pode ser considerada mapeando os ângulos necessários a sua implementação. Uma opção direta para a preparação de operadores unitário é a porta lógica

\begin{equation}
U(\theta, \phi, \gamma, \lambda) =
\begin{pmatrix}
e^{i\gamma} \cos(\theta/2) & -e^{i(\gamma + \lambda)} \cos(\theta/2) \\
e^{i(\gamma + \phi)} \cos(\theta/2) & e^{i(\gamma + \phi + \lambda)} \cos(\theta/2)
\end{pmatrix}.
\end{equation}

Fazendo $\theta = \pi/2, \phi = -\pi/2, \gamma = 3\pi/4, \lambda=\pi/2$, verifica-se que a correspondência com o operador $U$ definido pela forma de $U$ na base de $A$.Para $\theta = \pi, \phi = \pi, \gamma = 0, \lambda=0$, teremos então $U^2$. Portanto, a aplicação desses operadores no circuito se resume em  adicionar ângulos no circuito através da versão controlada da porta $U(\theta, \phi, \gamma, \lambda)$, conforme Boxe abaixo.

```python
qc.cu(np.pi/2, -np.pi/2, np.pi/2, 3*np.pi/4, qr_clock1, qr_b[0])
qc.cu(np.pi, np.pi, 0, 0, qr_clock2, qr_b[0])
qc.barrier()
```

Após aplicação das unitárias controladas, o estado é então escrito como

\begin{equation}
    \begin{split}
        \ket{\varphi_3} &=  \frac{1}{2\sqrt{2}}(-\ket{u_1}\ket{00} - i\ket{u_1}\ket{01} + \ket{u_1}\ket{10} + i\ket{u_1}\ket{11} \\
        &+\ket{u_2}\ket{00} - \ket{u_2}\ket{01} + \ket{u_2}\ket{10} - \ket{u_2}\ket{11}) \ket{0}_a
    \end{split}
\end{equation}

Em seguida, temos o uso da \textit{Inverse Quantum Fourier Transform}, a qual é responsável pela conversão da base de Fourier para a base computacional, cujo código pode ser visualizado no Boxe abaixo

```python
qc.h(qr_clock2)
qc.cp(-np.pi/2, qr_clock1, qr_clock2)
qc.h(qr_clock1)
qc.swap(qr_clock1, qr_clock2)
qc.barrier()
```

Após a aplicação da $QFT^{\dagger}$, teremos então o estado dado por

\begin{equation*}
    \ket{\varphi_4} = QFT^{\dagger} \ket{\varphi_3}
\end{equation*}

\begin{equation}
\begin{split}
    \ket{\varphi_4} = \frac{1}{\sqrt{2}}\left(-\ket{u_1}\ket{01} + \ket{u_2}\ket{10}\right)\ket{0}_a
\end{split}
\end{equation}

## Codificação Quântica de Ancilla

Aqui utilizamos rotações controladas considerando $R_Y(\theta)$, responsáveis por aumentar a probabilidade do qubit ancilla ser observado no estado $\ket{1}$. Há uma relação direta entre os autovalores decodificados e os ângulos, de modo que

\begin{equation}
    \theta (\tilde \lambda_1) = 2\arcsin{(1/1)} = \pi
\end{equation}

\begin{equation}
   \theta (\tilde \lambda_2) = 2\arcsin{(1/2)} = \pi/3
\end{equation}

De modo semelhante à etapa anterior, esses ângulos são substituídos em uma versão controlada da porta $R_Y$, conforme mostrado abaixo.

```python
qc.cry(np.pi, qr_clock1, qr_ancilla[0])
qc.cry(np.pi/3, qr_clock2, qr_ancilla[0])
qc.barrier()
```

Além disso, esta é também a etapa em que o qubit ancilla é medido, tendo o seu resultado armazenado no registrador clássico. Essa etapa resulta em

\begin{equation}
    \begin{split}
        \ket{\varphi_6} &= \sqrt{\frac{8}{5}} \left(-\frac{1}{\sqrt{2}}\ket{u_1}\ket{01}\ket{1}_a \right.\\
        & \left. + \frac{1}{2\sqrt{2}}\ket{u_2}\ket{10}\ket{1}_a\right)
    \end{split}
\end{equation}

Neste estágio, teremos então um estado próximo do esperado, mas impossível de ser medido diretamente na base computacional, visto que o estado de interesse $\ket{x}$ está emaranhado com os \textit{c-qubits}. 

## Estimativa de Fase Quântica Inversa

Neste estágio, aplicaremos a $QPE^{\dagger}$ para possibilitar a medida do estado $\ket{x}$ na base computacional de forma correta. Para tanto, a QFT é aplicada visando a codificação da informação em frequência, conforme explicitado no Boxe abaixo.

```python
qc.swap(qr_clock1, qr_clock2)
qc.h(qr_clock1)
qc.cp(np.pi/2, qr_clock1, qr_clock2)
qc.h(qr_clock2)
qc.barrier()
qc.measure(qr_ancilla, cl[ancilla:])
qc.barrier()
```

Teremos então o estado

\begin{equation}
    \begin{split}
        \ket{\varphi_7} &= \sqrt{\frac{8}{5}}\left(-\frac{1}{\sqrt{2}}\ket{u_1}\frac{1}{2}\left(\ket{00} + i\ket{01} - \ket{10} 
        \right. \right.\\
        &\left. -i\ket{11}\right)\ket{1}_a + \frac{1}{2\sqrt{2}}\ket{u_2}\frac{1}{2}\left(\ket{00} - \ket{01} \right.\\
        &\left. \left.
        + \ket{10} - \ket{11}\right)\right)\ket{1}_a
    \end{split}
\end{equation}

Em seguida faremos o uso da evolução inversa ($e^{-iAt}$), o que restaura o estado inicial. Nesse ponto, as matrizes inversas podem ser calculadas seguindo a mesma lógica da aplicação das unitárias, mas considerando $U^{-l} = VU_d^{-l}V$. Com isso, teremos

\begin{equation}
    U^{-1} = \begin{pmatrix}
    -1-i & 1-i \\
    1-i &  -1-i \\
    \end{pmatrix}
\end{equation}

Que pode ser simplesmente aplicada fazendo $\theta = \pi/2$, $\phi = \pi/2$, $\gamma = -3\pi/4$ e $\lambda=-\pi/2$. Note que $U^2 = U^{-2}$, de modo que $\theta = \pi$, $\phi = \pi$, $\gamma = 0$ e $\lambda=0$. Essas operações são implementadas no box abaixo:

```python
qc.cu(np.pi, np.pi, 0, 0, qr_clock2, qr_b[0])
qc.cu(np.pi/2, np.pi/2, -np.pi/2, -3*np.pi/4, qr_clock1, qr_b[0])
qc.barrier()
```

Para a rotação controlada inversa o estado será multiplicado por $e^{-i\lambda_jt}$, $e^{-i2\lambda_jt}$ e $e^{-i3\lambda_jt}$. Visto que $e^{-i\lambda_1t}=-i$, $e^{-i2\lambda_1t}=-1$, $e^{-i3\lambda_1t}=i$ $e^{-i\lambda_2t}=-1$, $e^{-i2\lambda_2t}=1$ e $e^{-i3\lambda_2t}=-1$. Então teremos o estado dado por:

\begin{equation}
    \begin{split}
        \ket{\varphi_8} &= \sqrt{\frac{8}{5}}\left(-\frac{1}{\sqrt{2}}\ket{u_1}\frac{1}{2}(\ket{00} + \ket{01} + \ket{10} + \ket{11}\right)\ket{1}_a \\
        &+ \frac{1}{2\sqrt{2}}\ket{u_2}\frac{1}{2}\left(\ket{00} + \ket{01} + \ket{10} + \ket{11}\right)\ket{1}_a \\
        &= \frac{1}{2}\sqrt{\frac{8}{5}}\left(-\frac{1}{(1)\sqrt{2}}\ket{u_1}\frac{1}{(2)\sqrt{2}}\ket{u_2} \right)\left(\ket{00} +  \ket{01} \right.
        \\ 
        & \left. + \ket{10} + \ket{11}\right)\ket{1}_a 
    \end{split}
\end{equation}

Na última etapa utilizamos novamente a porta Hadamard, finalizando o circuito com a inversão da sobreposição uniformemente distribuída feita aos clock qubits durante a rotina QPE, e deve ser aplicado como demonstra o boxe abaixo.

```python
qc.h(qr_clock1)
qc.h(qr_clock2)
qc.barrier()
```

O que resulta no estado dado por

\begin{equation}
        \ket{\varphi_9} = \sqrt{\frac{8}{5}} \left( -\frac{1}{(1)\sqrt{2}}\ket{u_1}+\frac{1}{(2)\sqrt{2}}\ket{u_2} \right) \ket{00}\ket{1}_a
\end{equation}

Após a substituição na base computacional de $\ket{u_1}=\frac{1}{\sqrt{2}}\ket{0} - \frac{1}{\sqrt{2}}\ket{1}$ e $\ket{u_2}=\frac{1}{\sqrt{2}}\ket{0} + \frac{1}{\sqrt{2}}\ket{1}$, teremos.

\begin{equation}
    \ket{\varphi_9} = \left(-\frac{\sqrt{10}}{10}\ket{0} + \frac{3\sqrt{10}}{10}\ket{1}\right)\ket{00}\ket{1}_a    
\end{equation}

## Resultados numéricos

Após a compilação das rotinas necessárias para o real fun-
cionamento do algoritmo, é possível então avaliar a
qualidade dos seus resultados. Para tanto, devemos escolher a
arquitetura utilizada (simulador ou computador quântico real)
e a quantidade de medidas (shots) para, por fim, plotarmos a
distribuição de probabilidade da medida, conforme o boxe.
Nesse caso, o ambiente selecionado foi o AerSimulator(), um
simulador sem ruído baseado em medidas.

```python
qc.barrier()
qc.measure(qr_b, cl[:ancilla])

simulator = AerSimulator()
new_circuit = transpile(qc, simulator)
job = simulator.run(new_circuit, shots =4096)
result = job.result()
counts = result.get_counts()
total_shots = sum(counts.values())
probabilities = {state: count / total_shots for state, count in counts.items()}
plot_distribution(probabilities)
```
